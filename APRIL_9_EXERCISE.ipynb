## Step 2: Mathematical formulation of factor analysis

Factor analysis assumes that the observed data can be explained by a smaller number of hidden variables called **latent factors**.

A common matrix form of the factor model is:

$$
X = L F + \varepsilon
$$

where:

* $X$ is the matrix of **observed variables**
* $L$ is the **loading matrix**
* $F$ is the matrix of **factor scores**
* $\varepsilon$ is the **residual/error matrix**

In element-wise form, this can be written as:

$$
x_{ij} = \lambda_{j1} f_{i1} + \lambda_{j2} f_{i2} + \cdots + \lambda_{jk} f_{ik} + \varepsilon_{ij}
$$

or equivalently,

$$
x_{ij} = \sum_{m=1}^{k} f_{im}\lambda_{jm} + \varepsilon_{ij}
$$

## Interpretation of the symbols in our example

In this notebook:

* each **row** corresponds to one company/brand
* each **column** corresponds to one measured vehicle-related property

So the observed matrix $X$ is our company-by-feature table.

If we have:

* $n$ companies
* $p$ observed variables
* $k$ latent factors

then:
$$
X \in \mathbb{R}^{n \times p}, \qquad
F \in \mathbb{R}^{n \times k}, \qquad
L \in \mathbb{R}^{p \times k}, \qquad
\varepsilon \in \mathbb{R}^{n \times p}
$$


## In this dataset

We currently have:

* $n = 9$ companies
* $p = 8$ observed properties

Thus, our observed matrix \(X\) has dimension:

$$
X \in \mathbb{R}^{9 \times 8}
$$

The aim of factor analysis is to describe these 8 observed properties using a smaller number $k$ of latent factors.

For example, the analysis may reveal hidden factors such as:

1. luxury/performance
2. commercial utility
3. Private mobility
4. Two-wheeler mobility

These factors are not directly measured, but are inferred from the correlation structure of the observed variables, so our common factor space is $k=4$

```python
import pandas as pd

print("Original company-by-feature table:")
display(df)

factor_map = {
    1: ("luxury/performance", ["Luxury_Level", "Performance", "Price_Level"]),
    2: ("commercial_utility", ["Cargo_Capacity", "Commercial_Focus", "Passenger_Capacity"]),
    3: ("private_mobility", ["Passenger_Capacity", "Urban_Compactness", "Price_Level"]),
    4: ("two_wheeler_mobility", ["Two_Wheeler_Focus", "Urban_Compactness", "Price_Level"])
}

print("\nAvailable mobility preference options:")
print("1. Luxury / Performance")
print("2. Commercial Utility")
print("3. Private Mobility")
print("4. Two-Wheeler Mobility")

try:
    choice = int(input("\nEnter a number (1-4): ").strip())

    if choice in factor_map:
        factor_name, selected_features = factor_map[choice]

        print(f"\nSelected category: {factor_name}")
        print(f"Relevant features used: {selected_features}")

        result = df.copy()
        result["Score"] = result[selected_features].sum(axis=1)

        ranked = result.sort_values("Score", ascending=False)

        print("\nRanked companies based on the selected mobility factor:")
        display(ranked[selected_features + ["Score"]])

    else:
        print("Invalid option. Please enter a number from 1 to 4.")

except ValueError:
    print("Invalid input. Please enter an integer from 1 to 4.")
```
